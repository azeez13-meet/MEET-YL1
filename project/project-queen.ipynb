```python
import csv
f = open('melanoma.csv', 'rb')
reader = csv.reader(f)
index = 0
Queen = []
Jennifer = []
for row in reader:
    if index != 0:
        r = map(float, row)
        Queen.append(r[0])
        Jennifer.append(r[1])
        print r
    index += 1
f.close()
```

```python
from pylab import *
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Cancer victims in the past years')
ax.set_xlabel('the years')
ax.set_ylabel('how many times the incidence happend ')
ax.plot(Queen, Jennifer)
fig.savefig("oyster_with_titles.png")
```
